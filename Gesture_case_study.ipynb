Problems with the Accelerometer data:

The accelerometer observations provided in the given case study is based on the data collected by the IMU or Inertial Measurement Unit Sensors. However, IMU sensor data is problematized by various experts as per our online research. One of the major IMUâ€™s disadvantages is their accumulated error when integrating them with navigation equations alone in order to find the position. For example, double integration of the noisy acceleration leads to an inaccurate position solution that eventually results in a huge drift. Hence, many tracking navigation applications use additional sensors to mitigate these effects. Please refer to this link: https://towardsdatascience.com/what-is-imu-9565e55b44c. 

In addition, it is difficult to represent gestures in three dimensions as human gestures recorded by sensors contain considerable noise and are uncertain because of the possibility of orientations of gestures made by the users at different angles. The gesture recognition becomes all the more challenging because we have observed in the dataset that the length of each of the sequences of x,y, and z coordinates that together make a gesture vary widely even across repetitions of same gesture 20 times by the same user and also between users. 

It is also important to note that the dictionary of 12 hand gestures mentioned in the publication given along with the case study assignment question does not take into account whether those 12 gestures were performed by the users using their left hand or right hand. There are also issues in making more accurate gesture recognition as the gestures differ in shape and in the lengths of sequence of three-dimensional coordinates that constitute a gesture. For instance, two gestures in the dictionary of 12 hand gestures identified as 'S' and 'AS' in the given publication signify the trajectory when the hand is moved to the right and trajectory when the hand is moved to the left respectively. These two gestures, 'S' and 'AS' each represent a circular trajectory in three-dimensional space and hence the the starting points as well as the diameters of those circular trajectories can vary if performed by the same accelerometer user 20 times or performed by each of the 8 users. Researchers have failed to specify in the given publication what methods they have used to claim recognition of a complete gesture accurately with considerable possibility of different starting points and diameters of a circular trajectories in 'S' and 'AS' hand gestures. 

To summarise there are slew of major problems with accelerometer data when it comes to gesture recognition as investigated by the researchers in the given publication and the dataset that is given to us in the case study:

1. Different accelerometer sensor users can perform the same hand gesture differently.

2. Even the same accelerometer sensor user does not perform the same gesture in exactly in same manner every time he repeats a particular gesture. 

3. We have observed by looking at the different lengths of sequences of movements that constitute a gesture that gesture recognition is also subject to the varied behavior of users in performing hand gestures as recorded by accelerometer data. 

4. Most importantly, we are not aware of the source of the dataset that is given to us in the case study with respect to the device or sensor used to generate the particular accelerometer readings as we stated earlier that there are problems if IMU sensors are used in research study. We are also not aware whether the researchers, who generated the data given to us in the case study used a template of standard gestures to match the patterns and identify the occurrence of the 20 hand gestures in the dataset. 

5. The given dataset only contains information about the coordinates without any information about the orientations in space as we pointed out earlier considering the users performing hand gestures at different angles.Furthermore, we don't know which of the hand gestures' three-dimensional coordinates were captured by the accelerometer using the left hand alone, right hand alone or both the hands of the users.

6. We also are not aware of the category of each of the 8 users, whose hand gestures were recorded in the dataset. For instance, if few of the users are kids, who performed hand gestures, then the researchers would keep the confidence level low and if other users, who performed the hand gestures are adults, then the researchers, who prepared the given dataset would keep a different confidence level. Here, 'Confidence' pertains to the probability of a particular hand movement or frame that is part of a particular gesture is the same as the one that was intended by the researchers to happen when they were conducting their research study. 

Our online research about recommended approach for gesture recognition in accelerometer data:

We reviewed the research literature available online and concluded that the implementation of statistical approach, Hidden Markov Models is highly recommended and performed by many researchers for gesture recognition using accelerometer data and we have attached the links to refer to those articles. HMM(Hidden Markov Models) is recommended because it considers sequential data such as different frames or movements that together constitute a gesture and are claimed to perform better than most machine learning algorithms when it comes to gesture recognition from accelerometer data. Thus, Hidden Markov Models as a statistical approach is very relevant to the given case study. However, we were unable to retrieve a suitable Python code for Hidden Markov Models implementation based on our online investigation of research literature, official documentation of 'hmmlearn' Python package and other documents and support our analysis.
